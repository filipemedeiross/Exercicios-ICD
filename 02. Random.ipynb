## Python - Random - Exercício

Implemente a função `generate_password` que deve gerar uma senha aleatória.

A função tem tres parâmetros:

- `n_chars` especifica o tamanho da senha gerada
- `use_numbers` especifica se números podem ser usados na senha. Por padrão, esse parâmetro tem valor True
- `use_special_chars` especifica se caracteres especiais, por exemplo !#%&@?, pode ser usados na senha. Por padrão, esse parâmetro tem valor True

Note que se `use_numbers` e `use_special_chars` tiverem valor False, a senha terá apenas letras maiúsculas ou minúsculas.

```python
import random

def generate_password(n_chars, use_numbers=True, use_special_chars=True):
    
    use_chars = [dec_ascii for dec_ascii in list(range(65, 91)) + list(range(97,123))]
    
    if use_numbers:
        use_chars.extend([dec_ascii for dec_ascii in list(range(48, 58))])
    if use_special_chars:
        use_chars.extend([dec_ascii for dec_ascii in list(range(33, 48)) + list(range(58, 65))
                          + list(range(91, 97)) + list(range(123, 127))])
        
    password = [chr(dec_ascii) for dec_ascii in random.sample(use_chars, n_chars)]
    
    return ''.join(password)
```

```python
print(generate_password(10))
print(generate_password(10, use_numbers=False))
print(generate_password(10, use_numbers=False, use_special_chars=False))
```

```python
# Opção alternativa de resposta

import random

def select_char(*args):
    return random.choice([random.randrange(interval[0], interval[1]) for interval in args[0]])

def generate_password2(n_chars, use_numbers=True, use_special_chars=True):
    
    use_chars = [(65, 91), (97,123)]
    
    if use_numbers:
        use_chars.append((48, 58))
    if use_special_chars:
        use_chars.extend([(33, 48), (58, 65), (91, 97), (123, 127)])
        
    password = [chr(select_char(use_chars)) for x in range(n_chars)]
    
    return ''.join(password)

print(generate_password2(10))
print(generate_password2(10, use_numbers=False))
print(generate_password2(10, use_numbers=False, use_special_chars=False))
```
